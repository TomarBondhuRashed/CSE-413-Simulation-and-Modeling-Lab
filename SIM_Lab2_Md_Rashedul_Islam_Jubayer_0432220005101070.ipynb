#Student's Assignment 1: Basic Matrix Operations

```python
import numpy as np

# Create two 3Ã—3 matrices with random integers between 1 and 10
A = np.random.randint(1, 11, size=(3, 3))
B = np.random.randint(1, 11, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix Addition
add_result = A + B
print("\nAddition:\n", add_result)

# Matrix Subtraction
sub_result = A - B
print("\nSubtraction:\n", sub_result)

# Matrix Multiplication
mul_result = A @ B
print("\nMultiplication:\n", mul_result)

# Determinant of A
det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

# Inverse of A (if invertible)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:\n", inv_A)
else:
    print("\nMatrix A is not invertible (det = 0)")

# Rank of A
rank_A = np.linalg.matrix_rank(A)
print("\nRank of A:", rank_A)
```

#Student's Assignment 2: Matrix Plotting and Visualization

**1.Generate and Plot Two Random Vectors:**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Part 1: Generate and Plot Two Random Vectors
vec1 = np.random.rand(15)
vec2 = np.random.rand(15)

plt.figure(figsize=(8, 4))
plt.plot(vec1, label='A', marker='o')
plt.plot(vec2, label='B', marker='s')
plt.title("Random Vectors Comparison")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

**2.Generate and Visualize a Random Matrix:**

```python
# Part 2: Generate and Visualize a Random 4x4 Matrix
matrix_4x4 = np.random.rand(4, 4)

plt.figure(figsize=(6, 5))
sns.heatmap(matrix_4x4, annot=True, cmap='viridis', xticklabels=[1,2,3,4], yticklabels=[1,2,3,4])
plt.title("4x4 Random Matrix Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

**3.Matrix Operations and Visualization:**

```python
# Part 3: Matrix Operations and Visualization
M1 = np.random.rand(4, 4)
M2 = np.random.rand(4, 4)

# Addition
add_result = M1 + M2

# Subtraction
sub_result = M1 - M2

# Multiplication
mul_result = M1 @ M2

# Flatten each result for bar plot
def plot_matrix_bar(matrix, title):
    flat = matrix.flatten()
    plt.figure(figsize=(6, 4))
    plt.bar(range(len(flat)), flat)
    plt.title(title)
    plt.xlabel("Element Index")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()

plot_matrix_bar(add_result, "Matrix Addition Result")
plot_matrix_bar(sub_result, "Matrix Subtraction Result")
plot_matrix_bar(mul_result, "Matrix Multiplication Result")
```
